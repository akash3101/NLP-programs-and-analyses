```python
performance = ["battery", "gaming", "windows", "light", "display", "office", "processor", "backup", 
               "weight", "camera", "speed", "keyboard", "sound", "video", "graphics", "intel", "games", 
               "heating", "editing", "online", "power", "installed", "features", "charging", "specs", 
               "ports", "update", "videos", "speakers", "lightweight", "processing", 
               "hanging", "motherboard", "screen", "microsoft", "charger", "software", "configuration", 
               "keypad", "webcam", "booting", "browsing", "resolution", "functioning", "fingerprint",
               "noise", "durability", "trackpad", "audio", "webcam", "flickering", "troubleshooting", 
               "draining", "adapter", "mouse", "retina", "boots", "pentium"]

customer = ["amazon", "service", "warranty", "months", "issue", "problem", "customer", "return", 
            "replacement", "waste", "received", "defective", "replaced", "center", "experience", 
            "seller", "replace", "detective", "option", "piece", "centre", "brand", "device", 
            "delivered", "mentioned", "repair", "selling", "recommend", "request", "model", 
            "damage", "exchange", "technician", "refund", "fault", "serial", "customers", "ordered", 
            "delivery", "store", "available", "student", "refused", 
            "invoice", "complaint", "cheating", "responding", "complaints", "supports"]

pricelooks = ["price", "light", "design", "build", "range", "looks", "plastic", "worth", 
              "smooth", "cheap", "built", "sleek", "metal", "buying", "sturdy", "colour", 
              "stylish", "beast", "premium", "beautiful", "packaging", "friendly", "portable", 
              "money", "display", "value", "heavy", "looking", "lightweight", "budget", "expensive", 
              "color", "costly", "discount", "white", "offer", "elegant"]
```

```python
p_revs = []
c_revs = []
l_revs = []
for review in unigrams_list:
    if len(set(review).intersection(set(performance))) > 0 and review not in p_revs:
        p_revs.append(review)
    if len(set(review).intersection(set(customer))) > 0 and review not in c_revs:
        c_revs.append(review)
    if len(set(review).intersection(set(pricelooks))) > 0 and review not in l_revs:
        l_revs.append(review)
```

```python
P = []
C = []
L = [] 
for review in unigrams_list:
    if review in p_revs:
        P.append(1)
    elif review not in p_revs:
        P.append(0)
    
    if review in c_revs:
        C.append(1)
    elif review not in c_revs:
        C.append(0)
        
    if review in l_revs:
        L.append(1)
    elif review not in l_revs:
        L.append(0)
```

```python
DD['performance'] = P
DD['customers'] = C
DD['pricelooks'] = L
```
